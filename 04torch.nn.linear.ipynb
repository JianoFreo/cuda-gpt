```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
torch.manual_seed(42)

torch.__version__
```

torch.nn.Linear(in_features = 3, out_features = 1, bias=True)

## formula = y = xW^T + b

### y = output
### x = input
### W = weight
### b = bias
### T = transpose



## torch.nnlinear architechture : y = xw^T + b

```python
# core code
# Create a simple Linear layer
linear = nn.Linear(in_features=10, out_features=5)

# Create input data
batch_size = 32
input_data = torch.randn(batch_size, 10)

# Forward pass
output = linear(input_data)

# Verify shapes
assert input_data.shape == (32, 10), f"Input shape: {input_data.shape}"
assert output.shape == (32, 5), f"Output shape: {output. shape}"

print(f"Input shape: {input_data.shape}")
print(f"Output shape: {output.shape}")
print(f"Weight shape: {linear.weight.shape}")
print(f"Bias shape: {linear.bias.shape}")
```
