```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([i for i in range(100)])
x = x.reshape(-1, 1) # reshape to make it a 2D array with one column and 100 rows
# -1 means "infer the number of rows based on the number of columns and the total number of elements"
# 1 means "1 column"
y = 46 + 2 * x.flatten() # flatten to convert the 2D array back to a 1D array
# 0 x 2 + 46 = 46
# 1 x 2 + 46 = 48
# 2 x 2 + 46 = 50
# ...
# 99 x 2 + 46 = 244
x,y
```

```python
plt.scatter(x, y, label="Initial data")
plt.title('Pre Pytorch')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Normalize data
normalized means we want to scale the data to a specific range. This is important because it can help improve the performance of machine learning models. For example, if we have a dataset with features that are on different scales, it can be difficult for the model to learn from the data. By normalizing the data, we can ensure that all features are on the same scale, which can help the model learn more effectively. 
## turn this to tensor

```python
array = np.array([i for i in range(100)])
x = array.reshape(-1, 1)
y = 46 + 2 * x.flatten() 

x_mean = x.mean()
x_std = x.std()

x_normalized = (x - x_mean) / x_std
x_tensor = torch.tensor(x_normalized, dtype=torch.float32)

y_normalized = (y - y.mean()) / y.std()
y_tensor = torch.tensor(y_normalized, dtype=torch.float32)

print(x_tensor.shape)
print(y_tensor.shape)
```

```python
plt.scatter(x_tensor, y_tensor, label="Normalized data")
plt.title('Normalized data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
