```python
randint = torch.randint(-100,100, (6,))
randint
# torch.randint(low, high, size)
# -100 — minimum value (inclusive)
# 100 — maximum value (exclusive), so the highest number you'll get is 99
# (6,) — shape of the tensor, meaning a 1D tensor with 6 random integers
```
